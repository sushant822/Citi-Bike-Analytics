```python
import pandas as pd
```

```python
df_2020_1 = pd.read_csv('data/2020/JC-202001-citibike-tripdata.csv')
df_2020_1.head()
```

```python
df_2020_2 = pd.read_csv('data/2020/JC-202002-citibike-tripdata.csv')
df_2020_3 = pd.read_csv('data/2020/JC-202003-citibike-tripdata.csv')
df_2020_4 = pd.read_csv('data/2020/JC-202004-citibike-tripdata.csv')
df_2020_5 = pd.read_csv('data/2020/JC-202005-citibike-tripdata.csv')
df_2020_6 = pd.read_csv('data/2020/JC-202006-citibike-tripdata.csv')
df_2020_7 = pd.read_csv('data/2020/JC-202007-citibike-tripdata.csv')
df_2020_8 = pd.read_csv('data/2020/JC-202008-citibike-tripdata.csv')
```
